````python
summary_table = pd.DataFrame()
summary_table['education'] = df['education']

summary_table['capital.diff'] = df['capital.gain'] - df['capital.loss']

summary_table = summary_table[summary_table['capital.diff'] != 0]

summary_table['categories'] = pd.qcut(summary_table['capital.diff'], q=10)

pivot_table = summary_table.pivot_table(index='categories', columns='education', aggfunc='size', fill_value=0)
print(pivot_table)
````

15) Создайте случайную колонку - magic_salary, которую нужно будет вычислить следующим образом: если зарплата небольшая (<50K), тогда случайно выберите число из диапазона [0,50]. Если зарплата выше 50K тогда из диапазона [51, 200]. Посчитайте среднюю зарплату в час для групп людей с одни уровнем образования на основе нашей случайной колонки magic_salary

````python
import numpy as np

df['magic_salary'] = np.where(df['income'] == ' <50K',
                              np.random.randint(0, 51, size=len(df)),
                              np.random.randint(51, 201, size=len(df)))

df['hourly_salary'] = df['magic_salary'] / (df['hours.per.week'] * 52)

average_hourly_salary_by_education = df.groupby('education')['hourly_salary'].mean()

print('Средняя зарплата в час для групп людей с одним уровнем образования:\n\n', average_hourly_salary_by_education)

average_magic_salary = df.groupby('education')['magic_salary'].mean()

print('\nСредняя зарплата в год для групп людей с одним уровнем образования:\n\n', average_magic_salary)
````

### 2. Корреляционная матрица

Напомним, что корреляция отражает взаимосвязь двух случайных величин. Она бывает положительная и отрицательная. Чем ближе коэффициент корреляции к нулю, тем меньше взаимосвязь. Чем больше абсолютная корреляци, тем взаимосвязь больше.

#### 9. Постройте plot.pie, показывающий распределение студентов по оценкам

Оценки студентов выставляются по шкале:<br>
0  - 40 marks : grade E<br>
41 - 60 marks : grade D<br>
60 - 70 marks : grade C<br>
70 - 80 marks : grade B<br>
80 - 90 marks : grade A<br>
90 - 100 marks : grade O<br>

Для этого посчитайте сумму результатов за 3 экзамена и найдите среднее. Оценка выставляется по среднему значению. Если студент не сдал математику(даже если средний балл выше 40), он получает Е

````python
def assign_grade(row):
    if row['math score'] < 40:
        return 'grade E'
    average_score = row['average score']
    if average_score < 40:
        return 'grade E'
    elif 40 <= average_score < 61:
        return 'grade D'
    elif 61 <= average_score < 71:
        return 'grade C'
    elif 71 <= average_score < 81:
        return 'grade B'
    elif 81 <= average_score < 91:
        return 'grade A'
    else:
        return 'grade O'

df['math score'] = df['math score'].replace('Сдал', 100)
df['reading score'] = df['reading score'].replace('Сдал', 100)
df['writing score'] = df['writing score'].replace('Сдал', 100)
df['math score'] = df['math score'].replace('Не сдал', 0)
df['reading score'] = df['reading score'].replace('Не сдал', 0)
df['writing score'] = df['writing score'].replace('Не сдал', 0)

df['average score'] = (pd.to_numeric(df['math score']) + pd.to_numeric(df['reading score']) + pd.to_numeric(df['writing score'])) / 3

df['summary grade'] = df.apply(assign_grade, axis=1)
summary_grade_counts = df['summary grade'].value_counts(sort=False)
# Перечисляем все оценки в правильном порядке
grades_order = ['grade E', 'grade D', 'grade C', 'grade B', 'grade A', 'grade O']

plt.figure(figsize=(8, 8))
summary_grade_counts.plot.pie(autopct='%1.1f%%', startangle=90, labels=summary_grade_counts.index, counterclock=False, wedgeprops={'edgecolor': 'gray'})
plt.title('Распределение студентов по оценкам')
plt.ylabel('')
plt.show()

````
