```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt

def equat(x):
  return x ** 3 + np.cos(x)

x0_bis = (-1, 1)
x0_new = -1
x0_sec = (-1, 1)

sol_bis = root_scalar(equat, method='bisect', bracket=x0_bis)
sol_new = root_scalar(equat, method='newton', x0=x0_new)
sol_sec = root_scalar(equat, method='secant', x0=x0_sec[0], x1=x0_sec[1])

x_val = np.linspace(-2, 2, 500)
y_val = equat(x_val)

plt.plot(x_val, y_val, label='x³+cos(x)')
plt.axhline(0, color='black', linewidth=0.75)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x)=x³+cos(x)')
plt.grid(True)
plt.legend()
plt.show()

print('Корени уравнения методом бисекции: ', sol_bis.root)
print('Корени уравнения методом Ньютона: ', sol_new.root)
print('Корени уравнения методом секущих: ', sol_sec.root)
```

**Ход работы:**

1. **Определение переменных**: Определяем переменные $ \mu $, $ \nu $, и $ \gamma $, которые будут использоваться в наших уравнениях.

3. **Создание массивов данных**: Создаем два массива: `x_values` и `y_values`. `x_values` содержит значения $ x $, которые мы получаем, добавляя к $ \mu + \nu $ числа от 0 до 8 с шагом 2. `y_values` содержит соответствующие значения $ y $, которые мы также вычисляем, используя заданные значения $ \gamma $ и добавляя к нему разные числа.

4. **Подготовка матрицы коэффициентов**: Создаем матрицу `A`, которая будет использоваться для нахождения коэффициентов нашей параболы. Каждая строка матрицы `A` представляет собой значения $ 1 $, $ x $, и $ x^2 $ для каждого значения из `x_values`.

5. **Решение методом наименьших квадратов**: Функция `lstsq` принимает нашу матрицу `A` и массив `y_values`, и находит коэффициенты $ c_0 $, $ c_1 $, и $ c_2 $, которые минимизируют разницу между фактическими значениями $ y $ и значениями, предсказанными нашей параболой.

6. **Вывод результатов**: В конце мы выводим найденные коэффициенты и сумму квадратов отклонений, которая показывает, насколько хорошо наша парабола соответствует данным.

```python
import numpy as np
from scipy.linalg import lstsq
import matplotlib.pyplot as plt

mu = 2
nu = 3
gamma = 1
thetta = 2

x_values = np.array([mu + nu + i*2 for i in range(5)])
y_values = np.array([gamma + thetta + 2.5,
                     gamma + thetta + 4.9,
                     gamma + thetta + 8,
                     gamma + thetta + 12.1,
                     gamma + thetta + 16.9])

A = np.vstack([np.ones_like(x_values), x_values, x_values**2]).T
coefs, residues, _ , _ = lstsq(A, y_values)

print(f"Коэффициенты параболы c0={coefs[0]}, c1={coefs[1]}, c2={coefs[2]}")
print("Сумма квадратов отклонений:", residues)

x_range = np.linspace(x_values.min(), x_values.max(), 100)
y_parabola = coefs[0] + coefs[1]*x_range + coefs[2]*x_range**2

plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, label='Исходные данные')
plt.plot(x_range, y_parabola, color='red', label='Аппроксимирующая парабола y=c0+c1x+c2x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Аппроксимация данных параболой y=c0+c1x+c2x²')
plt.legend()
plt.grid(True)
plt.show()

A_exp = np.vstack([np.ones_like(x_values), np.log(x_values)]).T

coefs_exp, residues_exp, _, _ = lstsq(A_exp, y_values)

print(f"Коэффициенты экспоненциальной кривой a={np.exp(coefs_exp[0])}, b={np.exp(coefs_exp[1])}")
print("Сумма квадратов отклонений для экспоненциальной аппроксимации:", residues_exp)

y_exp = np.exp(coefs_exp[0]) * np.exp(coefs_exp[1])**x_range

plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, label='Исходные данные')
plt.plot(x_range, y_parabola, color='red', label='Аппроксимирующая парабола')
plt.plot(x_range, y_exp, color='green', label='Аппроксимирующая экспоненциальная кривая y=ab^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Аппроксимация данных y=ab^x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def f_a(x, y, nu, mu, thetta, gamma):
  return y + (nu / x)

def f_b(x, y, nu, mu, thetta, gamma):
  return (y + x**2 * np.cos(x)) / x

def euler_method(f, x0, y0, nu, mu, theta, gamma, h):
  x_values = [x0]
  y_values = [y0]
  x = x0
  y = y0
  while len(x_values) < 4:
      y += h * f(x, y, nu, mu, theta, gamma)
      x += h
      x_values.append(x)
      y_values.append(y)
  return x_values, y_values

def runge_kutta_method(f, x0, y0, nu, mu, theta, gamma, h):
  x_values = [x0]
  y_values = [y0]
  x = x0
  y = y0
  while len(x_values) < 4:
      k1 = h * f(x, y, nu, mu, theta, gamma)
      k2 = h * f(x + h, y + k1, nu, mu, theta, gamma)
      y += 0.5 * (k1 + k2)
      x += h
      x_values.append(x)
      y_values.append(y)
  return x_values, y_values

nu = 0.1
mu = 0.2
theta = 0.3
gamma = 0.4
x0_a = 3 + nu + 0.1*mu
y0_a = 2 + theta + 0.1*gamma
x0_b = 0.5 * (mu + theta + 1)
y0_b = (nu + gamma) / 2
h=0.2

x_val_euler_a, y_val_euler_a = euler_method(f_a, x0_a, y0_a, nu, mu, theta, gamma, h)
x_val_euler_a = np.round(x_val_euler_a, 2)
y_val_euler_a = np.round(y_val_euler_a, 2)
x_val_euler_b, y_val_euler_b = euler_method(f_b, x0_b, y0_b, nu, mu, theta, gamma, h)
x_val_euler_b = np.round(x_val_euler_b, 2)
y_val_euler_b = np.round(y_val_euler_b, 2)

x_val_rk_a, y_val_rk_a = runge_kutta_method(f_a, x0_a, y0_a, nu, mu, theta, gamma, h)
x_val_rk_a = np.round(x_val_rk_a, 2)
y_val_rk_a = np.round(y_val_rk_a, 2)
x_val_rk_b, y_val_rk_b = runge_kutta_method(f_b, x0_b, y0_b, nu, mu, theta, gamma, h)
x_val_rk_b = np.round(x_val_rk_b, 2)
y_val_rk_b = np.round(y_val_rk_b, 2)

def ode_func_a(x, y):
  return y + (nu / x)

def ode_func_b(x, y):
  return (y + x**2 * np.cos(x)) / x

sol_a = solve_ivp(ode_func_a, (x0_a, x0_a + 2), [y0_a], dense_output=True)
sol_b = solve_ivp(ode_func_b, (x0_b, x0_b + 2), [y0_b], dense_output=True)

x_val_plot_a = np.linspace(x0_a, x0_a + 2, 100)
y_val_plot_a = sol_a.sol(x_val_plot_a).flatten()

x_val_plot_b = np.linspace(x0_b, x0_b + 2, 100)
y_val_plot_b = sol_b.sol(x_val_plot_b).flatten()

print('№\tМетод Эйлера\t\tМетод Рунге-Кутта\n')
for i in range(4):
  print(f'a)\tx{i+1}={x_val_euler_a[i]} y{i+1}={y_val_euler_a[i]}\t\tx{i+1}={x_val_rk_a[i]} y{i+1}={y_val_rk_a[i]}')
print('\n')
for i in range(4):
  print(f'б)\tx{i+1}={x_val_euler_b[i]} y{i+1}={y_val_euler_b[i]}\t\tx{i+1}={x_val_rk_b[i]} y{i+1}={y_val_rk_b[i]}')
print('\n')

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x_val_plot_a, y_val_plot_a, label='solve_ivp', color='black')
plt.plot(x_val_euler_a, y_val_euler_a, 'o--', label='Метод Эйлера')
plt.plot(x_val_rk_a, y_val_rk_a, 's--', label='Метод Рунге-Кутта')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Решение ОДУ (а)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_val_plot_b, y_val_plot_b, label='solve_ivp', color='black')
plt.plot(x_val_euler_b, y_val_euler_b, 'o--', label='Метод Эйлера')
plt.plot(x_val_rk_b, y_val_rk_b, 's--', label='Метод Рунге-Кутта')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Решение ОДУ (б)')
plt.legend()

plt.tight_layout()
plt.show()
```
